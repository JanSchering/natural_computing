```python
import matplotlib.pyplot as plt
from tabulate import tabulate
```

(Selection Pressure) [1 point] Given the fitness function $f(x) = x^2$, calculate the
probability of selecting the individuals $x = 2$, $x = 3$, and $x = 4$, using roulette wheel
selection. Calculate the probability of selecting the same individuals when the fitness
function is scaled as follows $f_1(x) = f(x) + 20$.

```python
# prepare functions
def f(x):
    return x**2

def f_scaled(x):
    return f(x) + 20

x1 = 2
x2 = 3
x3 = 4
x_names = ["x1", "x2", "x3"]
x_vals = [x1, x2, x3]
```

(a) For each of the two fitness functions: 1) provide a table with fitness function
value and selection probability for each individual; 2) plot the pie chart (roulette
wheel) with selection probability of the three individuals.


```python
# create Table and plot for f(x)

# get fitnesses using f(x)
f_x1 = f(x1)
f_x2 = f(x2)
f_x3 = f(x3)
f_vals = [f_x1, f_x2, f_x3]

# get summed_fitness
sum_f = sum(f_vals)

# calc selection probabilities
p1 = f_x1/sum_f
p2 = f_x2/sum_f
p3 = f_x3/sum_f
p_vals = [p1, p2, p3]

# Create pie chart
labels = x_names
values = p_vals
fig, ax = plt.subplots(2, facecolor='w')
ax[0].pie(values, labels=labels, autopct='%1.1f%%', startangle=90)


# create table
headers = ["value", "f(x)", "selection prob."]
ax[1].set_axis_off() 
table = ax[1].table( 
    cellText = [[x_vals[i], f_vals[i], p_vals[i]] for i in range(len(x_vals))],  
    rowLabels = x_names,  
    colLabels = headers, 
    rowColours =["palegreen"] * 10,  
    colColours =["palegreen"] * 10, 
    cellLoc ='center',  
    loc ='upper left')

fig.suptitle("Using f(x)=x^2 to get selection probabilities")
fig.set_dpi(150)
```

```python
# create Table and plot for f_scaled(x)

# get fitnesses
f_vals = [f_scaled(x1), f_scaled(x2), f_scaled(x3)]

# get summed_fitness
sum_f = sum(f_vals)

# calc selection probabilities
p_vals = [f_vals[i]/sum_f for i in range(len(x_vals))]

# Create pie chart
labels = x_names
values = p_vals
fig, ax = plt.subplots(2, facecolor='w')
ax[0].pie(values, labels=labels, autopct='%1.1f%%', startangle=90)

# create table
headers = ["value", "f(x)", "selection prob."]
ax[1].set_axis_off() 
table = ax[1].table( 
    cellText = [[x_vals[i], f_vals[i], p_vals[i]] for i in range(len(x_vals))],  
    rowLabels = x_names,  
    colLabels = headers, 
    rowColours =["palegreen"] * 10,  
    colColours =["palegreen"] * 10, 
    cellLoc ='center',  
    loc ='upper left')

fig.suptitle("Using f(x)=x^2 + 20 to get selection probabilities")
fig.set_dpi(150)



```

(b) Which fitness function yields a lower selection pressure?

The selection pressure is lower when using $f_1(x)=x^2+20$. The reason for this is that the variance of the individuals' fitnesses using the fitness function $f(x) = x^2$ is $\sigma ^2 = 36.333$ whilst it is equal to $\sigma ^2 = 6.333$ when using the fitness function $f_1(x) = x^2 + 20$. This decrease in variance is an indication that the selection probabilities for the different values are much closer to each other than for the unscaled fitness function, since the selection probabilities are computed as such: selection probability of individual $x_i$ is equal to $\frac{f_i}{\Sigma f}$. With selection probabilities of individuals that are more similar to each other, the fitness function $f_1(x)$ has decreased the selection pressure as compared to using the fitness function $f(x)$.

 

(c) What can you conclude about the effect of fitness scaling on selection pressure?

By applying a fitness scaling, the discrepancy in the selection probabilities between different individuals decreases. We can conclude that fitness scaling is an effective measure to adjust the selection pressure on the population. The higher we scale the fitness, the less fitness actually matters. Fitness scaling can completely remove selective pressure as the effect of the perceived performance becomes negligible in relation to the scale.

Example: performances in range [0, 10] getting scaled by a factor of $10e^5$. If $x_1=10$ and $x_2=0$, the selection probability $p(x_1)=0.49999976$ while $p(x_2)=0.5000025$. Thus, despite the gap in performance their probability is the same with six digit precision.
